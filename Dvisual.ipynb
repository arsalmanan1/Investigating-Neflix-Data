# DATA VISUALISATION

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
boat = sns.load_dataset('titanic')
boat.head()
```

```python
plt.plot(boat['age'],boat['fare'])
plt.show
```

```python
# customise
plt.plot(boat['age'],boat['fare'],color='red' )
plt.xlabel('age of paseengers')
plt.ylabel('fare')
plt.title('age - fare relatons')
plt.show
```

```python
# customise and scatter plot
plt.plot(boat['age'],boat['fare'], 'o' )
plt.xlabel('age of paseengers')
plt.ylabel('fare')
plt.title('age - fare relatons')
plt.show
```

```python
# customise and scatter
plt.scatter(boat['age'],boat['fare'],color='green' )
plt.xlabel('age of paseengers')
plt.ylabel('fare')
plt.title('age - fare relatons')
plt.show
```

```python
colors = []
for sex in boat['sex']:
    if sex == 'male':
        colors.append('blue')
    else:
        colors.append('pink')

plt.scatter(boat['age'],boat['fare'],c= colors )
plt.xlabel('age of paseengers')
plt.ylabel('fare')
plt.title('age - fare relatons')
plt.show
```

```python
colors = ['red' if sex == 'male'else 'pink' for sex in boat['sex']]
plt.scatter(boat['age'],boat['fare'],c= colors )
plt.xlabel('age of paseengers')
plt.ylabel('fare')
plt.title('age - fare relatons')
plt.show
```

```python
# histogram
plt.hist(boat['age'] )
plt.xlabel('age of paseengers')
plt.show
```

```python
#value count
boat['sex'].value_counts().sum()
```
